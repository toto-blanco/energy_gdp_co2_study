```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("indus_pourc_pib.csv")
```

```python
df
```

```python
#suppression colonne Indicator Code et Indicator Name
df=df.drop(["Indicator Name","Indicator Code"],axis=1)
print(df)
```

```python
# Supprimer tout ce qui suit la virgule dans la colonne 'Country Name'
df.loc[:, 'Country Name'] = df['Country Name'].str.split(',').str[0].str.strip()

# Obtenir la liste des valeurs uniques de Country Name
countries = df['Country Name'].unique()

# Afficher toutes les valeurs
print(f"Nombre total de pays : {len(countries)}\n")
print("Liste complète :")
for country in sorted(countries):
    print(country)
```

```python
#transformation des colonnes de 1960 à 2017 en ligne en créant la colonne year et la colonne ratio indus/pib en utilisant fonction melt

# création d'une liste des colonnes d'années
annees = [str(annee) for annee in range(1960,2017)]

#transformation des colonens d'années en ligne avec melt
df_trans = df.melt(
    id_vars = ["Country Name","Country Code"], # colonne à garder
    value_vars = annees,                      # colonne à transformer
    var_name = "Year",                        # nom nouvelle colonne pour les années
    value_name = "ratio Indus/PIB"            # nom nouvelle colonne pour valeurs               
)
                              
# convertir la colonne year en entier
df_trans["Year"] = df_trans["Year"].astype(int)

print(df_trans)

```

```python
#suppression des lignes de df_trans qui ont un iso_code = NaN

df_trans = df_trans.dropna(subset=["Country Code"])

with pd.option_context('display.max_rows', None):
    print(df_trans)
```

```python
# Vérification

# Charger liste de référence des codes ISO 3166-1 alpha-3
iso_codes = pd.read_csv("slim-3.csv")["alpha-3"].tolist()   

df_trans_codes = df_trans["Country Code"].unique()

# Chercher codes non valides
invalid_codes = [code for code in df_trans_codes if code not in iso_codes]

# Afficher un message en fonction du résultat
if not invalid_codes:
    print("✅ Tous les codes pays sont valides (ISO 3166-1 alpha-3).")
else:
    print(f"❌ Codes non valides détectés : {invalid_codes}")
    print("Voici les lignes concernées :")
    df_invalid = df_trans[df_trans["Country Code"].isin(invalid_codes)]
    print(df_invalid)
```

```python
#détermination du nombre de valeur NaN et pourcentage de valeur Nan par pays

#définir nombre de colonne (hors country)
total_cols = len(df_trans.columns) - 1

#calcul du nombre de NaN par pays
nan_counts = df_trans.isna().sum(axis=1) #compte les NaN par ligne
df_nan = df_trans.assign(nan_count = nan_counts) #ajout d'une colonne temporaire avec le nombre de Nan de la ligne

#grouper par pays et calculer la somme du nombre de Nan et le pourcentage de NaN
result = (
    df_nan.groupby("Country Name")
    .agg(
        total_nan = ("nan_count","sum"),
        ratio_nan = ("nan_count",lambda x: (x.sum() / (len(x) * total_cols)) * 100)
    )
    .reset_index()
)

# ordonner result par ratio desc
result = result.sort_values(by='ratio_nan', ascending=False)

with pd.option_context('display.max_rows', None):
    print(result)
```

```python
# ordonner result par ratio desc
result = result.sort_values(by='Country Name', ascending=True)

with pd.option_context('display.max_rows', None):
    print(result)
```
