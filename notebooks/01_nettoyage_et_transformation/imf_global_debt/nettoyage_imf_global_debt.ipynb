```python
# Supprimer tout ce qui suit la virgule dans la colonne 'COUNTRY'
df.loc[:, 'COUNTRY'] = df['COUNTRY'].str.split(',').str[0].str.strip()

# Obtenir la liste des valeurs uniques de COUNTRY
countries = df['COUNTRY'].unique()

# Afficher tous les pays
print(f"Nombre total de pays : {len(countries)}\n")
print("Liste complète :")
for country in sorted(countries):
    print(country)

```

```python
# filtrer et ne garder que les lignes qui ont pour INDICATOR : "Gross domestic product (GDP), Current prices, Domestic currency", "Debt instruments, Public sector, Percent of GDP", "Debt instruments, Private sector, Percent of GDP" 


# Liste des indicateurs à conserver
indicators_to_keep = [
    "Debt instruments, General government, Percent of GDP",
    "Debt instruments, Private sector, Percent of GDP"
]

# Filtrer le DataFrame pour ne garder que ces lignes
df_filtered = df[df['INDICATOR'].isin(indicators_to_keep)]

# Afficher le résultat
print(df_filtered.head())

```

```python
#création fonction diagnostic
def diagnostic_df(df):
    info_list = []
    for col in df.columns:
        data = df[col]
        info = {
            "colonne": col,
            "dtype": data.dtype,
            "nb_valeurs_uniques": data.nunique(dropna=True),
            "nb_valeurs_manquantes": data.isna().sum(),
            "%_valeurs_manquantes": round(data.isna().mean() * 100, 2),
            "exemple_valeur": data.dropna().iloc[0] if data.dropna().shape[0] > 0 else None,
        }
        if np.issubdtype(data.dtype, np.number):
            info.update({
                "min": round(data.min(), 2),
                "max": round(data.max(), 2),
                "moyenne": round(data.mean(), 2),
                "médiane": round(data.median(), 2)
            })
        else:
            info.update({"min": None, "max": None, "moyenne": None, "médiane": None})
        info_list.append(info)
    return pd.DataFrame(info_list).sort_values(by="%_valeurs_manquantes", ascending=False)
```

```python
diagnostic_df(df_filtered)
```

```python
#détermination du nombre de valeur NaN et pourcentage de valeur Nan par pays

#définir nombre de colonne (hors country)
total_cols = len(df_filtered.columns) - 1

#calcul du nombre de NaN par pays
nan_counts = df_filtered.isna().sum(axis=1) #compte les NaN par ligne
df_filtered_nan = df_filtered.assign(nan_count = nan_counts) #ajout d'une colonne temporaire avec le nombre de Nan de la ligne

#grouper par pays et calculer la somme du nombre de Nan et le pourcentage de NaN
result = (
    df_filtered_nan.groupby("COUNTRY")
    .agg(
        total_nan = ("nan_count","sum"),
        ratio_nan = ("nan_count",lambda x: (x.sum() / (len(x) * total_cols)) * 100)
    )
    .reset_index()
)

# ordonner result par ratior desc

result = result.sort_values(by='ratio_nan', ascending=False)

with pd.option_context('display.max_rows', None):
    print(result)
```
