```python
import pandas as pd

# --- Liste des fichiers ---
files = {
    "energy": "/Users/amysarr/Documents/Formation_Jedha/Projet_final/fichier_comparaison/global-data-on-sustainable-energy_CLEAN.csv",
    "co2": "/Users/amysarr/Documents/Formation_Jedha/Projet_final/fichier_comparaison/data_co2_emissions_across_clean.csv",
    "owid": "/Users/amysarr/Documents/Formation_Jedha/Projet_final/fichier_comparaison/indus_pourc_pib_clean.csv",
    "indus": "/Users/amysarr/Documents/Formation_Jedha/Projet_final/fichier_comparaison/owid-energy-data_clean.csv",
    "imf": "/Users/amysarr/Documents/Formation_Jedha/Projet_final/fichier_comparaison/imf_global_debt_database_clean.csv",
    "pop":"/Users/amysarr/Documents/Formation_Jedha/Projet_final/fichier_comparaison/pourc_pop_urb_clean.csv"
}

# --- Nom de la colonne pays dans chaque fichier ---
country_columns = {
    "energy": "Entity",
    "co2": "Name",
    "indus": "country",     
    "owid": "Country Name",
    "imf": "COUNTRY",
    "pop": "Country Name"         
}

# --- Lecture fichiers ---
dfs = {name: pd.read_csv(path) for name, path in files.items()}

# --- Extraction des pays ---
countries = {
    name: set(df[country_columns[name]].dropna().unique())
    for name, df in dfs.items()
}

# --- Trouver les pays communs à tous ---
common_all = set.intersection(*countries.values())

# --- Trouver les pays absents dans certains fichiers ---
missing = {
    name: (common_all - country_set)
    for name, country_set in countries.items()
}

# --- Trouver les pays différents par fichier ---
different = {
    name: country_set - set.union(*(v for k, v in countries.items() if k != name))
    for name, country_set in countries.items()
}

# --- Affichage vertical ---

print("\n=========== LISTE DES PAYS COMMUNS ===========\n")
for country in sorted(common_all):
    print(country)

print("\n=========== PAYS MANQUANTS PAR FICHIER ===========\n")
for name, miss in missing.items():
    print(f"\n-- Pays absents dans {name} ({len(miss)}) --")
    for country in sorted(miss):
        print(country)

print("\n=========== PAYS DIFFÉRENTS PAR FICHIER ===========\n")
for name, diff in different.items():
    print(f"\n-- Pays uniques à {name} ({len(diff)}) --")
    for country in sorted(diff):
        print(country)

```
